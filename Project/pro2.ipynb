```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# --- PREPROCESSING (Run this first to load and prep data) ---
df = pd.read_csv('commodities_indian_units.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Price_Range'] = df['High'] - df['Low']
df['Daily_Change'] = df['Close'] - df['Open']

df.head()
```

```python
# Q1: Find and visualize the average Close price of each commodity.
avg_close = df.groupby('Commodity')['Close'].mean().reset_index()
display(avg_close.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=avg_close, x='Commodity', y='Close', palette='viridis')
plt.title('1. Average Close Price by Commodity')
plt.show()
```

```python
# Q2: Find and visualize the total Volume traded for each commodity.
total_vol = df.groupby('Commodity')['Volume'].sum().reset_index()
display(total_vol.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=total_vol, x='Commodity', y='Volume', palette='magma')
plt.title('2. Total Volume by Commodity')
plt.show()
```

```python
# Q3: Find the maximum High price recorded for each commodity.
max_high = df.groupby('Commodity')['High'].max().reset_index()
display(max_high.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=max_high, x='Commodity', y='High', palette='Reds')
plt.title('3. Maximum High Price by Commodity')
plt.show()
```

```python
# Q4: Find the minimum Low price recorded for each commodity.
min_low = df.groupby('Commodity')['Low'].min().reset_index()
display(min_low.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=min_low, x='Commodity', y='Low', palette='Blues')
plt.title('4. Minimum Low Price by Commodity')
plt.show()
```

```python
# Q5: Count the total number of records available for each commodity.
record_count = df['Commodity'].value_counts().reset_index()
record_count.columns = ['Commodity', 'Count']
display(record_count.head())
plt.figure(figsize=(6, 6))
plt.pie(record_count['Count'], labels=record_count['Commodity'], autopct='%1.1f%%')
plt.title('5. Record Count Distribution')
plt.show()
```

```python
# Q6: Find the average Open price per year across all commodities.
avg_open_year = df.groupby('Year')['Open'].mean().reset_index()
display(avg_open_year.head())
plt.figure(figsize=(8, 4))
sns.lineplot(data=avg_open_year, x='Year', y='Open', marker='o')
plt.title('6. Average Open Price Per Year')
plt.show()
```

```python
# Q7: Find the average price range (High - Low) by commodity.
avg_range = df.groupby('Commodity')['Price_Range'].mean().reset_index()
display(avg_range.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=avg_range, x='Commodity', y='Price_Range')
plt.title('7. Average Price Range by Commodity')
plt.show()
```

```python
# Q8: Show the yearly trend of the Close price for Gold.
gold_yearly = df[df['Commodity'] == 'Gold'].groupby('Year')['Close'].mean().reset_index()
display(gold_yearly.head())
plt.figure(figsize=(8, 4))
sns.lineplot(data=gold_yearly, x='Year', y='Close', marker='o', color='gold')
plt.title('8. Yearly Close Price Trend for Gold')
plt.show()
```

```python
# Q9: Plot the monthly Volume trend for Silver.
silver_monthly_vol = df[df['Commodity'] == 'Silver'].groupby('Month')['Volume'].mean().reset_index()
display(silver_monthly_vol.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=silver_monthly_vol, x='Month', y='Volume', color='silver')
plt.title('9. Monthly Volume Trend for Silver')
plt.show()
```

```python
# Q10: Show the trend of average Close price across all commodities.
yearly_close_all = df.groupby(['Year', 'Commodity'])['Close'].mean().reset_index()
display(yearly_close_all.head())
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly_close_all, x='Year', y='Close', hue='Commodity', marker='o')
plt.title('10. Yearly Close Price Trend by Commodity')
plt.show()
```

```python
# Q11: Plot the Open vs Close price trend for Silver over the years.
silver_open_close = df[df['Commodity'] == 'Silver'].groupby('Year')[['Open', 'Close']].mean().reset_index()
display(silver_open_close.head())
plt.figure(figsize=(8, 4))
plt.plot(silver_open_close['Year'], silver_open_close['Open'], label='Open', marker='o')
plt.plot(silver_open_close['Year'], silver_open_close['Close'], label='Close', marker='x')
plt.title('11. Open vs Close Price for Silver')
plt.legend()
plt.show()
```

```python
# Q12: Compare the yearly High price trend of Gold and Silver.
gold_silver_high = df[df['Commodity'].isin(['Gold', 'Silver'])].groupby(['Year', 'Commodity'])['High'].mean().reset_index()
display(gold_silver_high.head() )
plt.figure(figsize=(8, 4))
sns.lineplot(data=gold_silver_high, x='Year', y='High', hue='Commodity', marker='o')
plt.title('12. Yearly High Price: Gold vs Silver')
plt.show()
```

```python
# Q13: Compare the total historical Volume of Gold and Silver.
gs_vol = df[df['Commodity'].isin(['Gold', 'Silver'])].groupby('Commodity')['Volume'].sum().reset_index()
display(gs_vol)
plt.figure(figsize=(6, 4))
sns.barplot(data=gs_vol, x='Commodity', y='Volume', palette='Set2')
plt.title('13. Total Volume: Gold vs Silver')
plt.show()
```

```python
# Q14: Compare the Open price distribution between Gold and Crude Oil using a Boxplot.
gold_crude_open = df[df['Commodity'].isin(['Gold', 'Crude Oil'])][['Commodity', 'Open']]
display(gold_crude_open.groupby('Commodity').describe())
plt.figure(figsize=(8, 4))
sns.boxplot(data=gold_crude_open, x='Commodity', y='Open')
plt.title('14. Open Price Distribution: Gold vs Crude')
plt.show()
```

```python
# Q15: Compare the yearly Volume distribution using a stacked bar chart.
yearly_vol_pivot = df.pivot_table(index='Year', columns='Commodity', values='Volume', aggfunc='sum')
display(yearly_vol_pivot.head())
yearly_vol_pivot.plot(kind='bar', stacked=True, figsize=(10, 5), colormap='Set3')
plt.title('15. Yearly Volume Distribution Across Commodities')
plt.show()
```

```python
# Q16: Compare the overall price range (High-Low) across all commodities.
range_desc = df.groupby('Commodity')['Price_Range'].describe()
display(range_desc.head())
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Commodity', y='Price_Range')
plt.title('16. Price Range Distribution by Commodity')
plt.show()
```

```python
# Q17: Compare the Open vs Close prices for Crude Oil using a Scatter Plot.
crude_scatter = df[df['Commodity'] == 'Crude Oil'][['Open', 'Close']]
display(crude_scatter.head())
plt.figure(figsize=(6, 6))
sns.scatterplot(data=crude_scatter, x='Open', y='Close', color='darkred', alpha=0.5)
plt.title('17. Open vs Close for Crude Oil')
plt.show()
```

```python
# Q18: Find the median Close price of each commodity.
median_close = df.groupby('Commodity')['Close'].median().reset_index()
display(median_close.head())
plt.figure(figsize=(8, 4))
sns.barplot(data=median_close, x='Commodity', y='Close', palette='coolwarm')
plt.title('18. Median Close Price by Commodity')
plt.show()
```

```python
# Q19: Generate a correlation matrix of market metrics.
corr_matrix = df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()
display(corr_matrix)
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('19. Correlation Heatmap')
plt.show()
```

```python
# Q20: Analyze the Month-wise average Close price for Crude Oil.
crude_monthly_close = df[df['Commodity'] == 'Crude Oil'].groupby('Month')['Close'].mean().reset_index()
display(crude_monthly_close.head())
plt.figure(figsize=(8, 4))
sns.lineplot(data=crude_monthly_close, x='Month', y='Close', marker='s', color='brown')
plt.title('20. Month-wise Avg Close Price for Crude Oil')
plt.show()
```

```python
# Q21: Find the distribution of Close prices for Copper using a histogram.
copper_close = df[df['Commodity'] == 'Copper'][['Close']]
display(copper_close.describe())
plt.figure(figsize=(8, 4))
sns.histplot(data=copper_close, x='Close', bins=30, kde=True, color='orange')
plt.title('21. Distribution of Close Prices for Copper')
plt.show()
```

```python
# Q22: Show the yearly trend of the Low price for Gold.
gold_yearly_low = df[df['Commodity'] == 'Gold'].groupby('Year')['Low'].mean().reset_index()
display(gold_yearly_low.head())
plt.figure(figsize=(8, 4))
sns.lineplot(data=gold_yearly_low, x='Year', y='Low', marker='v', color='goldenrod')
plt.title('22. Yearly Low Price Trend for Gold')
plt.show()
```
